```python
import random

def generate_secret_number(length=4):
    secret_number = ''
    for _ in range(length):
        secret_number += str(random.randint(0, 9))
    return secret_number

def get_cows_bulls(secret_number, user_guess):
    cows, bulls = 0, 0
    for i, digit in enumerate(user_guess):
        if digit == secret_number[i]:
            cows += 1
        elif digit in secret_number:
            bulls += 1
    return cows, bulls

def play_cows_and_bulls():
    secret_number = generate_secret_number()
    attempts = 0

    print("Welcome to the Cows and Bulls game!")
    print("Try to guess the 4-digit secret number. For every correct digit in the correct position, you get a cow. For every correct digit in the wrong position, you get a bull.")
    print("Enter 'exit' to quit the game.")

    while True:
        user_guess = input("Enter your guess: ")

        if user_guess.lower() == 'exit':
            print("Thanks for playing! The secret number was:", secret_number)
            break

        if len(user_guess) != 4 or not user_guess.isdigit():
            print("Please enter a 4-digit number.")
            continue

        attempts += 1
        cows, bulls = get_cows_bulls(secret_number, user_guess)

        if cows == 4:
            print(f"Congratulations! You've guessed the secret number {secret_number} correctly in {attempts} attempts.")
            break
        else:
            print(f"Cows: {cows}, Bulls: {bulls}")

if __name__ == "__main__":
    play_cows_and_bulls()
```

```python
import itertools

def generate_all_numbers(length=4):
    return [''.join(digits) for digits in itertools.product('0123456789', repeat=length)]

def get_cows_bulls(secret_number, guess):
    cows, bulls = 0, 0
    for i, digit in enumerate(guess):
        if digit == secret_number[i]:
            cows += 1
        elif digit in secret_number:
            bulls += 1
    return cows, bulls

def solve_cows_and_bulls(secret_number):
    candidates = generate_all_numbers()
    attempts = 0

    while candidates:
        attempts += 1
        guess = candidates.pop()
        cows, bulls = get_cows_bulls(secret_number, guess)

        if cows == 4:
            return guess, attempts

        candidates = [candidate for candidate in candidates if get_cows_bulls(guess, candidate) == (cows, bulls)]

# Test the solver with a sample secret number
secret_number = "1234"
guess, attempts = solve_cows_and_bulls(secret_number)
print(f"The secret number {secret_number} was found in {attempts} attempts. Guessed number: {guess}")
```
