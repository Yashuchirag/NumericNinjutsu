```python
from fastai.collab import *
from fastai.tabular.all import *
```

```python
import pandas as pd

data_path = "path/to/your/data.csv"
df = pd.read_csv(data_path)
```

```python
dls = CollabDataLoaders.from_df(df, user_name='user_id', item_name='movie_id', rating_name='rating', bs=64)
```

```python
learn = collab_learner(dls, n_factors=50, y_range=(0, 5.5))
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(5, lr_max=1e-2)
```

```python
user_id = 1
movie_id = 100
input_data = {'user_id': user_id, 'movie_id': movie_id}
```

```python
pred = learn.predict(input_data)
```

```python
predicted_rating = float(pred[0])
```

```python
user_ids = [1, 2, 3, 4, 5]
movie_ids = [100, 101, 102, 103, 104]
input_df = pd.DataFrame({'user_id': user_ids, 'movie_id': movie_ids})
```
